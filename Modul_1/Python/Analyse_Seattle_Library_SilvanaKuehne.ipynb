```python
import urllib.request
import pandas as pd
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2018&$limit=10000"
checkout_file = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file)
```

```python
checkouts_df = pd.read_csv(checkout_file)
```

```python
checkouts_df
```

```python
#Wie sind die genauen Spaltennamen anzusprechen?
checkouts_df.columns
```

```python
#Welche Materialtypen gibt es?
checkouts_df["materialtype"].unique()
```

```python
#Gib mir alle Datensätze, die mit dem Materialtyp “Book” aus
checkouts_df[checkouts_df["materialtype"] == "BOOK"]
```

```python
#Mach daraus eine eigene Datei mit dem Namen checkouts_books_df
checkouts_books_df = checkouts_df[checkouts_df["materialtype"] == "BOOK"]
```

```python
#Gib mir aus der neuen Datei alle Datensätze, die über das System Horizon verbucht wurden
checkouts_books_df[checkouts_books_df ["checkouttype"] == "Horizon"]
```

```python
#Erstelle eine neue Datei in der die Filter Buch als Materialtyp und Checkoutsystem Horizon erfasst sind
checkouts_booksAndHorizon_df = checkouts_books_df[checkouts_books_df ["checkouttype"] == "Horizon"]
```

```python
#Wandel mir diese neue Datei in eine CSV um
checkouts_booksAndHorizon_df.to_csv("checkout_books_over_horizon.csv")
```

```python
#Gib mir alle Datensätze aus, deren Ausleihanzahl höher als 100 ist
checkouts_df[checkouts_df["checkouts"] > 100]
```

```python
#Zeige mir Zeilen- und Spaltenanzahl
checkouts_df[checkouts_df["checkouts"] > 100].shape
```

```python
#Gib mir die Abfrage in Diagrammform aus
checkouts_df[checkouts_df["checkouts"] > 100] ["checkouts"].plot(kind="bar")
```

```python
#Zeige mir, die Gesamtanzahl der erzielten Ausleihen pro Verlag
checkouts_df["publisher"].value_counts()
```

```python
#Zeige mir alle Ausleihen des Verlages Books on Tape
checkouts_df[checkouts_df["publisher"] == "Books on Tape"]
```

```python
#Erstelle eine neue Datei basierend auf den Kriterien  Verlag=Books on Tape
checkouts_publisher_df = checkouts_df[checkouts_df["publisher"] == "Books on Tape"]
```

```python
#Sortiere die Datei nach Anzahl der Ausleihen
checkouts_publisher_df.sort_values("checkouts")
```

```python
#Wandel die Datei von Books on Tape zu einer Excel um
checkouts_publisher_df.to_excel("checkout_publisher_BooksOnTape.xlsx")
```
